```python
lm.predict([[724,5575,1497,300,42,139,383,973,104,641,601,3.73,2,8,56,88]])
# to check if i can predict nearby y variable.
# so here i have considered the first data from x variable and predicted the y variable .Which had came out nearby 
# (Actual=95,Predicted=93.5)
```

# Looking for patterns in the residuals

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
#Returns the mean squared error; we'll take a square root
np.sqrt(mean_squared_error(y, y_pred))
```

```python
r_squared = r2_score(y, y_pred)
r_squared
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor


# the independent variables set
x

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = x.columns

# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(x.values, i)
                   for i in range(len(x.columns))]

print(vif_data)
```
